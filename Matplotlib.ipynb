# LinePlot

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 20, 1000)
```

```python
y = np.sin(x) + 0.2 * x
```

```python
plt.plot(x, y)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Title");
```

#  ScatterPlot

```python
x = np.random.randn(100, 2)
plt.scatter(x[:, 0], x[:, 1]);
```

```python
x = np.random.randn(200, 2)
x[: 100] += 3
y = np.zeros(200)
y[:  100] = 1
plt.scatter(x[:, 0], x[:, 1], c=y) 
```
